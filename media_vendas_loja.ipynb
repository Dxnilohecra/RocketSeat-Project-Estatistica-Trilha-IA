```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dict_faturamento = {
    'data_ref': [
        '2023-01-01',
        '2020-02-01',
        '2021-03-01',
        '2022-04-01',
        '2023-05-01',
        '2023-06-01',
        '2020-07-01',
        '2021-08-01',
        '2022-09-01',
        '2023-10-01',
        '2022-11-01',
        '2023-12-01',
        ],
    'valor': [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame(dict_faturamento)
df_faturamento
```

```python
df_faturamento.info()
```

```python
data_ref = pd.to_datetime(df_faturamento['data_ref'])
data_ref.info()
```

```python
data_ref = data_ref.sort_values()
data_ref = data_ref.reset_index(drop=True)
data_ref
```

```python
df_faturamento['data_ref'] = data_ref
df_faturamento.info()
```

Média de Vendas

```python
media_vendas = df_faturamento['valor'].mean()
media_vendas
```

Gráfico de Barras da Evolução de Faturamento

```python
df_faturamento.plot.bar(x='data_ref', y='valor')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Evolução de Faturamento')
plt.show()
```

Gráfico de Linhas da Evolução de Faturamento

```python
df_faturamento.plot(x='data_ref', y='valor')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Evolução de faturamento')
plt.show()
```
